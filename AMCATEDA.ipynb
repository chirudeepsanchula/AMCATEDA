# AMCAT DATA Exploratory Data Analysis


## Step 1 - Introduction

#### Objective: 

`The objective to perform exploratory data analysis (EDA) on AMCAT exam data is to understand the characteristics, patterns, and trends of the data, and to identify any errors, outliers, or anomalies that may affect the analysis. EDA can help to answer questions such as:`

- What are the distributions of the scores and salaries of the candidates?
- What are the correlations between the scores and salaries of the candidates?
- What are the factors that influence the scores and salaries of the candidates, such as gender, specialization, college tier, etc.?
- How do the scores and salaries of the candidates vary across different regions, states, or cities?
- Are there any significant differences or similarities between the scores and salaries of the candidates from different groups or categories?

`EDA can also help to prepare the data for further analysis, such as hypothesis testing, machine learning, or visualization. EDA can help to perform tasks such as:`

- Cleaning the data by removing or imputing missing values, correcting errors, or handling outliers
- Transforming the data by scaling, normalizing, encoding, or aggregating the variables
- Feature engineering by creating new variables from existing ones, or selecting the most relevant or important variables
- Data visualization by using plots, charts, graphs, or maps to display the data in a meaningful and appealing way.

#### Detailed overview: 

`The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate.`

## Step 5 - Research Questions
### Question 1: Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.

````python
cs_data = data[(data['Specialization'] == 'computer science & engineering') | (data['Specialization'] == 'computer engineering')]
cs_data = cs_data[(cs_data['Designation'] == 'programming analyst') | (cs_data['Designation'] == 'software engineer') | (cs_data['Designation'] == 'hardware engineer') | (cs_data['Designation'] == 'associate engineer')]
````

````python
cs_mean = cs_data['Salary'].mean()
cs_std = cs_data['Salary'].std()
print('The mean salary of the Computer Science Engineering graduates with the mentioned designations is', cs_mean)
````

````python
print('The standard deviation of the salary of the Computer Science Engineering graduates with the mentioned designations is', cs_std)
````

````python
claim_low = 250000
claim_high = 300000
if (claim_low <= cs_mean <= claim_high):
    print('The claim is true')
else:
    print('The claim is false')
````

### Observation: The claim is false, as the mean salary of the Computer Science Engineering graduates with the mentioned designations is higher than 3 lakhs.

````python
sns.boxplot(data=cs_data, x='Designation', y='Salary')
plt.xlabel('Designation')
plt.ylabel('Salary')
plt.title('Box plot of Salary by Designation')
plt.axhline(y=claim_low, color='red', linestyle='--', label='Claim low')
plt.axhline(y=claim_high, color='green', linestyle='--', label='Claim high')
plt.legend()
plt.show()
````

### Plot Observation: This plot shows that the median salary of each designation is above the claim range, confirming that the claim is false.

### Question 2: Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

````python
from scipy import stats
import statsmodels.api as sm
````

````python
contingency_table = pd.crosstab(data['Gender'], data['Specialization'])
````

````python
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print('The p-value of the chi-square test is', p)
alpha = 0.05
if (p < alpha):
    print('We reject the null hypothesis and conclude that there is a relationship between gender and specialization')
else:
    print('We fail to reject the null hypothesis and conclude that there is no relationship between gender and specialization')
````

### Observation: We reject the null hypothesis and conclude that there is a relationship between gender and specialization, as the p-value is very small.

````python
contingency_table = pd.crosstab(data['Gender'], data['Specialization'])
percentages = contingency_table.div(contingency_table.sum(axis=1), axis=0) * 100

percentages.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.title('Stacked bar plot of Gender and Specialization with Percentages')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
````

### Plot Observation: This plot shows the percentage of each specialization for each gender. We can see that there are some differences in the preference of specialization depending on the gender. For example, more males than females choose mechanical engineering, while more females than males choose biotechnology. We can also compare the percentages with the overall percentages of each specialization, which are shown by the dashed lines.
